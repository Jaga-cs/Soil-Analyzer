```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/DELL/Downloads/AgriGrow-main/data.csv')
```

```python
df
```

```python
n = 0
p = 0
k = 0
s = 0
oc = 0
ca = 0
score = 0

label = []
```

```python
for i in range(df.shape[0]):
    if(df.loc[i]['N']<=280):
        n=1
    elif(df.loc[i]['N']>280 and df.loc[i]['N']<=560):
        n=2
    elif(df.loc[i]['N']>560):
        n=3
        
        
    
    if(df.loc[i]['P']<=10):
        p=1
    elif(df.loc[i]['P']>10 and df.loc[i]['P']<=24.6):
        p=2
    elif(df.loc[i]['P']>24.6):
        p=3
        
        
        
    if(df.loc[i]['K']<=117.6):
        k=1
    elif(df.loc[i]['K']>117.6 and df.loc[i]['K']<=280):
        k=2
    elif(df.loc[i]['K']>280):
        k=3
        
    if(df.loc[i]['S']<=10):
        s=1
    elif(df.loc[i]['S']>10 and df.loc[i]['S']<=20):
        s=2
    elif(df.loc[i]['S']>20):
        s=3
        
    if(df.loc[i]['OC']<=0.5):
        oc=3
    elif(df.loc[i]['OC']>0.5 and df.loc[i]['OC']<=0.75):
        oc=6
    elif(df.loc[i]['OC']>0.75):
        oc=9
        
    if(df.loc[i]['Ca']<=2500):
        ca=1
    elif(df.loc[i]['Ca']>2500 and df.loc[i]['Ca']<=4000):
        ca=2
    elif(df.loc[i]['Ca']>4000):
        ca=3   
        
    #score = ((n+p+k+s+oc+ca)*10)/28.35
    score = ((n+p+k+s+oc+ca)*10)/24
    
    label.append((score))
    
```

```python
my_formatted_label = [ '%.3f' % elem for elem in label ]

sadf = pd.DataFrame(my_formatted_label)

sadf.columns = ['SFI'] 

sadf
```

```python
sadf = pd.concat((df, sadf), axis=1)
sadf.columns
```

```python
sadf=sadf.drop(['District'], axis=1)
SA_data = sadf.to_csv('SA_Data.csv', index = True)
```
